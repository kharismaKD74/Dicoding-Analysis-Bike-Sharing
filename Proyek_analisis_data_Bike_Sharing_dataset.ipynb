### Menggabungkan dua dataset menjadi satu dataframe

# Exploratory Data Analysis (EDA)

```python
# mendefinisikan fungsi yang akan digunakan untuk EDA
def range(series):
    return series.max() - series.min()
```

```python
bike_sharing.describe(include="all")
```

## Explore 

```python
bike_sharing.groupby(by="season_hourly").agg({
    "workingday_hourly": "count", #menjumlahkan penyewa pada hari kerja berdasarkan musim
    "windspeed_hourly": ["max", "min", "mean", lambda x: x.max() - x.min()]  # Membuat fungsi kustom untuk range
}).sort_values(by=("workingday_hourly", "count"), ascending=False)
```

```python
bike_sharing.groupby(by="season_daily").agg({
    "workingday_daily": "count", #menjumlahkan penyewa pada hari kerja berdasarkan musim
    "windspeed_daily": ["max", "min", "mean", lambda x: x.max() - x.min()]  # Membuat fungsi kustom untuk range
}).sort_values(by=("workingday_daily", "count"), ascending=False)
```

```python
group_season_month = bike_sharing.groupby(by="season_daily").mnth_daily.nunique()
group_season_month
```

```python
seasonal_data_daily = bike_sharing.groupby('season_daily')['cnt_daily'].count()
seasonal_data_daily
```

```python
weathersit_data_daily = bike_sharing.groupby('weathersit_daily')['cnt_daily'].count()
weathersit_data_daily
```

```python
working_data = bike_sharing.groupby('workingday_daily')['cnt_daily'].count().reset_index()
working_data
```

```python
working_data.loc[working_data["workingday_daily"] == 0, "workingday_daily"] = "Holiday"
working_data.loc[working_data["workingday_daily"] == 1, "workingday_daily"] = "Working Day"
working_data.rename(columns={"cnt_daily": "sum"}, inplace=True)
working_data
```

## Corelation
Melihat korelasi dari setiap data perhari

```python
numerical_columns = ["holiday_daily" 
 ,"weekday_daily","workingday_daily","weathersit_daily","temp_daily","atemp_daily","season_daily", "windspeed_daily",  "cnt_daily"]
correlation = bike_sharing[numerical_columns].corr()
print(correlation)
```

# Visualization & Explanatory Analysis

## 1. Apakah ada perbedaan antara hari kerja (workingday) dan hari libur (holiday) dalam jumlah sewa sepeda harian? 

```python
# Perbandingan hari kerja dan hari libur
#day_names = ['Holiday', 'Working day']
plt.bar(working_data["workingday_daily"], working_data["sum"])
plt.xlabel('Kategori Hari')
plt.ylabel('Jumlah Sewa Harian')
plt.title('Pengaruh Hari Kerja dan Holiday Terhadap Jumlah Sewa Sepeda Harian')
plt.show()
```

## 2. Bagaimana hubungan antara musim dan jumlah sewa sepeda harian ?

```python
# Pengaruh Musim
season_names = ['Spring', 'Summer', 'Fall', 'Winter']
plt.bar(season_names, seasonal_data_daily)
plt.xlabel('Musim')
plt.ylabel('Jumlah Sewa Harian')
plt.title('Pengaruh Musim Terhadap Jumlah Sewa Sepeda Harian')
plt.show()
```

## 3. Bagaimana pengaruh cuaca (weathersit) terhadap jumlah sewa sepeda harian? 

```python
# Pengaruh Weathersit
weathersit_names = ['Clear', 'Mist + Cloudy', 'Light Snow']
plt.bar(weathersit_names, weathersit_data_daily)
plt.xlabel('Cuaca')
plt.ylabel('Jumlah Sewa Harian')
plt.title('Pengaruh Cuaca Terhadap Jumlah Sewa Sepeda Harian')
plt.show()
```

## 4. Apakah ada pola berdasarkan waktu (bulan, jam) dalam jumlah sewa sepeda harian? 

```python
# Pola berdasarkan bulan
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(x="mnth_daily", y="cnt_daily", data=bike_sharing, ci=None)
plt.title("Pola Jumlah Sewa Sepeda Harian Berdasarkan Bulan")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Sewa Sepeda Harian")
plt.show()
```

```python
# Pola berdasarkan jam
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(x="hr", y="cnt_hourly", data=bike_sharing, ci=None)
plt.title("Pola Jumlah Sewa Sepeda Harian Berdasarkan Jam")
plt.xlabel("Jam")
plt.ylabel("Jumlah Sewa Sepeda Harian")
plt.show()
```

> Dilihat dari grafik diatas, sewa sepeda harian lebih banyak penggunaanya pada musim gugur (fall).

### 3. Bagaimana pengaruh cuaca terhadap jumlah sewa sepeda harian?
> Dilihat dari grafik diatas, cuaca sangat berpengaruh terhadap jumlah sewa sepeda harian. Sewa sepeda harian banyak digunakan pada cuaca cerah "clear", karena pada saat kondisi cuaca yang lain seperti berawan, hujan, ataupun bersalju kebanyakan orang pasti akan memilih transportasi lain untuk digunakan.

### 4. Apakah ada pola berdasarkan waktu (bulan, jam )dalam jumlah sewa sepeda harian?
> Iya, dapat dilihat dari grafik diatas, pada bulan 6 dan bulan 9 banyak yang menyewa sepeda dan berdassarkan jam dan jumlah sewa sepeda meningkat sekitar jam 8 pagi dan sekitar jam 5 atau 6 sore.
